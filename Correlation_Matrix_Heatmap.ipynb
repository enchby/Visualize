# 相関ヒートマップ

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```
#ダミーデータ生成

df = pd.DataFrame({
        'A' : [2, 2, 2, 3, 2, 4, 4],
        'B' : [1, 1, 2, 2, 3, 3, 4],
        'C' : [4, 1, 3, 2, 6, 3, 4],
        'D' : [2, 1, 3, 2, 4, 3, 4],
        'E' : [5, 4, 3, 1, 4, 3, 4],
        'F' : [2, 2, 5, 4, 1, 1, 4],
        'G' : [5, 3, 3, 2, 6, 4, 4]
    })

df
```

```
plt.figure(figsize=(12, 12))

#相関係数の計算
df = df.corr()

#重複する部分を非表示にするmaskの作成
mask = np.zeros_like(df, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True


sns.heatmap(df, mask = mask, annot=True, square = True, linewidths = .5, cmap = "BuPu")
plt.show()
```

### | r | = 相関係数

| r | =　0.7～1　　かなり強い正の相関がある

| r | =　0.4～0.7　正の相関あり

| r | =　0.2～0.4　弱い正の相関あり

| r | =　0～0.2　　ほとんど相関なし

| r | =　0～-0.2　　ほとんど相関なし

| r | =　-0.2～-0.4　弱い負の相関あり

| r | =　-0.4～-0.7　負の相関あり

| r | =　-0.7～-1　　かなり強い負の相関がある
