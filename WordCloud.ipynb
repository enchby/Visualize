```
import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
from wordcloud import WordCloud
from bs4 import BeautifulSoup  
from janome.tokenizer import Tokenizer
from janome.analyzer import Analyzer
#from janome.charfilter import *
from janome.tokenfilter import CompoundNounFilter
from collections import Counter
%matplotlib inline



#textファイルを読み込む場合
# text_file = open("something.txt")
# full_text = text_file.read()
# full_text= full_text.replace("\n","")



####### janomeによる形態素解析 start #######

#ノーマルTokenizer
# t = Tokenizer()
# tokens = t.tokenize(full_text)

#CompoundNounFilterを利用する際
a = Analyzer(token_filters=[CompoundNounFilter()])
tokens = a.analyze(text)

word_list=[]
for token in tokens:
    word = token.surface
    partOfSpeech = token.part_of_speech.split(',')[0]
    partOfSpeech2 = token.part_of_speech.split(',')[1]
     
    if partOfSpeech == "名詞":
        if (partOfSpeech2 != "非自立") and (partOfSpeech2 != "代名詞") and (partOfSpeech2 != "数"):
            word_list.append(word)

            
#もし、複数のテキストを解析してまとめるならば、ここでまとめる。ex. word_list = np.array(list_a) + np.array(list_b)
words_wakati=" ".join(word_list)

####### janomeによる形態素解析 end #######


#単語数の確認
noun_cnt = Counter(word_list)
print("Total words: ",sum(noun_cnt.values()))
print("Most common word: ",noun_cnt.most_common(5))



####### WordCloud実行 start #######


stop_words = [] 
    
fpath = "/usr/share/fonts/truetype/fonts-japanese-gothic.ttf"
#colaboratoryで他言語実行の場合は、下記pathにフォントをアップロードする
#fpath = "/usr/share/fonts/truetype/language.ttf"

wordcloud = WordCloud(
    font_path=fpath,
    width=1200, height=800,   # default width=400, height=200
    background_color="white",   # default=”black”
    stopwords=set(stop_words),
    max_words=500,   # default=200
    min_font_size=4,   #default=4
    collocations = False   #default = True
    ).generate(words_wakati)

####### WordCloud実行 end #######


print("Number of words used: ",len(wordcloud.words_))
# ファイルに保存する。
#wordcloud.to_file('wordcloud.png')
# numpy 配列で取得する。
img = wordcloud.to_array()


plt.figure(figsize=(12, 9))
plt.axis("off")  
plt.imshow(img)  
plt.show()

```
